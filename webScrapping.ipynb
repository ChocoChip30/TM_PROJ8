```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
import numpy as np
```

```python
def get_title(soup):
    try:
        title = soup.find("span", attrs={"id":'productTitle'})
        title_value = title.text.strip()
    except AttributeError:
        title_value = ""
    return title_value

def get_ingredients(soup):
    try:
        ingredients = soup.find("div", attrs={'id':'nic-ingredients-content'}).text.strip()
    except AttributeError:
        ingredients = ""
    return ingredients

def get_price(soup):
    try:
        price = soup.find("span", attrs={'class':'a-price a-text-price a-size-medium apexPriceToPay'}).text.strip()
    except AttributeError:
        price = ""
    return price
```

```python
if __name__ == '__main__':
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}
    url ="https://www.amazon.co.uk/s?k=gluten+free+snacks&crid=2G1IQMWSJ30CR&sprefix=gluten+free+snacks%2Caps%2C271&ref=nb_sb_noss_1"
    webpage = requests.get(url, headers=headers)

    soup = BeautifulSoup(webpage.content, "html.parser")
    links = soup.find_all("a", attrs={'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
    link_list = []

    for link in links:
        link_list.append(link.get('href'))

    d = {"title":[], "price":[], "ingredients":[], 'url':[]}

    tot = len(link_list)
    i = 0
    for link in link_list:
        l = "https://amazon.co.uk" + link
        each_page = requests.get(l, headers=headers)
        each_soup = BeautifulSoup(each_page.content, "html.parser")
        d['title'].append(get_title(each_soup))
        d['price'].append(get_price(each_soup))
        d['ingredients'].append(get_ingredients(each_soup))
        d['url'].append(l)
        print(i , "/" ,tot ," done")
        i += 1

   
    
```

```python
amazon_df = pd.DataFrame.from_dict(d)
amazon_df
```
